```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
# Task 1
```

```python
df = pd.read_csv( 'data/iris.csv' )
print( df.head( ) )
```

```python
y = df[ 'species' ]
df.drop( columns = [ 'species' ], inplace = True )
df_standardized = pd.DataFrame( StandardScaler( ).fit_transform( df ), columns = df.columns )
```

```python
print( df_standardized.head( ) )
```

```python
# Task 2
```

```python
kmeans_object = KMeans( n_clusters = 3, random_state = 1, init = 'k-means++', n_init = 10, copy_x = True, max_iter = 300, tol = 0.0001, verbose = 0 )
```

```python
kmeans_object.fit( df_standardized )
```

```python
# Task 3
```

```python
plotColors = [ 'orange', 'green', 'blue' ]
i=0
for species in y.drop_duplicates( ):
    mask = y == species
    plt.scatter( df_standardized[ mask ][ 'fsepal_length' ], df_standardized[ mask ][ 'sepal_width' ], c = plotColors[ i ], label = species )
    i+=1
plt.legend( y.drop_duplicates( ), title = 'species', loc = 'upper right' )
plt.xlabel( 'sepal_length' )
plt.ylabel( 'sepal_width' )
plt.show( )
```
